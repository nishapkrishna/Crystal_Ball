```python
from sklearn import tree
import pandas as pd
import os
import pickle
```

```python
df = pd.read_csv(os.path.join("..", "Resources", "movies_final.csv"))
df.head()
```

```python
data = df.drop("Profitable", axis=1)
feature_names = data.columns
data.head()
```

```python
target = df['Profitable']
# target_names = ["loss", "low", "medium", "high"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
# clf = tree.DecisionTreeClassifier()
# clf = clf.fit(X_train, y_train)
# clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, criterion='entropy')
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
feature_imp = sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
