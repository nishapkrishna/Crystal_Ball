```python
from sklearn import tree
import pandas as pd
import os
import pickle
```

```python
df = pd.read_csv(os.path.join("..", "Resources", "movies_final.csv"))
df.head()
```

```python
data = df.drop("Profitable", axis=1)
feature_names = data.columns
data.head()
```

```python
target = df['Profitable']
# target_names = ["loss", "low", "medium", "high"]
```

```python
from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=1, stratify=target)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
#train the model
classifier.fit(X_train, y_train)
```

```python
#test the model
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Dump the trained decision tree classifier with Pickle
lr_pkl_filename = '../Resources/logistic_regression_classifier_movie.pkl'
# Open the file to save as pkl file
lr_model_pkl = open(lr_pkl_filename, 'wb')
pickle.dump(classifier, lr_model_pkl)
# Close the pickle instances
lr_model_pkl.close()
```
