```python
import warnings
warnings.simplefilter('ignore')

# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
# Read the csv file into a pandas DataFrame

movie_genres = pd.read_csv('../Resources/movies_genres.csv')
movie_genres.head()
```

```python
data = movie_genres.copy()

movie_genres_binary_encoded = pd.get_dummies(data, columns=["genres_name"])
movie_genres_binary_encoded.head()
```

```python
movie_genres_binary_encoded_sum = movie_genres_binary_encoded.groupby('movie_id').sum()
```

```python
movie_genres_binary_encoded_sum.head()
```

```python
# movie_genres_binary_encoded_sum.count()
```

```python
# Read the csv file into a pandas DataFrame
movie_directors = pd.read_csv('../Resources/movies_directors.csv')
movie_directors.head()
```

```python
data = movie_directors.copy()
movie_directors_binary_encoded = pd.get_dummies(data, columns=["crew_name"])
movie_directors_binary_encoded.head()
```

```python
movie_directors_binary_encoded_sum = movie_directors_binary_encoded.groupby('movie_id').sum()
```

```python
movie_directors_binary_encoded_sum.head()
```

```python
movies_merged = pd.merge(movie_genres_binary_encoded_sum, movie_directors_binary_encoded_sum, on='movie_id')
```

```python
movies_merged.head()
```

```python
# Read the csv file into a pandas DataFrame
movie_actors = pd.read_csv('../Resources/movies_actors.csv')
movie_actors.head()
```

```python
data = movie_actors.copy()
movie_actors_binary_encoded = pd.get_dummies(data, columns=["actor_name"])
movie_actors_binary_encoded.head()
```

```python
movie_actors_binary_encoded_sum = movie_actors_binary_encoded.groupby('movie_id').sum()
```

```python
movie_actors_binary_encoded_sum.head()
```

```python
movies_merged_1 = pd.merge(movies_merged, movie_actors_binary_encoded_sum, on='movie_id')
movies_merged_1.head()
```

```python
# Read the csv file into a pandas DataFrame
movie_finance = pd.read_csv('../Resources/movies_profit_categorization.csv')
# movie_finance = movie_finance.set_index('movie_id')
movie_finance.head()
```

```python
data = movie_finance.copy()
movie_finance_binary_encoded = pd.get_dummies(data, columns=["budgetbin"])
movie_finance_binary_encoded.head()
```

```python
movie_profitability = movie_finance_binary_encoded.filter(['movie_id','budgetbin_0','budgetbin_1','budgetbin_2','Profitable'], axis=1)
movie_profitability = movie_profitability.set_index('movie_id')
movie_profitability.head()
```

```python
movies_merged_2 = pd.merge(movies_merged_1, movie_profitability, on='movie_id')
movies_merged_2.head()
```

```python
movies_merged_2.to_csv("../Resources/movies_final.csv", index=False)
```
